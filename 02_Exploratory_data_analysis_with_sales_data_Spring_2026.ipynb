```python
best_selling = coffee_data.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
print(best_selling)
import matplotlib.pyplot as plt
best_selling.plot(kind='bar')
plt.title('Total Revenue by Coffee Type')
```

# c)weekly revenue analysis

```python
week_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
revenue_by_day = coffee_data.groupby('Weekday')['money'].sum()
revenue_by_day = revenue_by_day.reindex(week_order)
print(revenue_by_day)
revenue_by_day.plot(kind = 'bar')
plt.title("Revenue by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Total Revenue")
```

**Analysis:** **Revenue by Weekday**

The weekly revenue analysis shows that **Tuesday** generates the **highest** revenue, while **Sunday** records the **lowest** sales.
Overall, revenue during weekdays appears stronger than weekends.
This suggests that coffee sales may be driven primarily by working-day customers, such as office-goers, rather than weekend leisure demand.

# f)monthly revenue trend(to smooth out the previous graph)


```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
coffee_data = pd.read_csv("Coffe_sales.csv")
coffee_data['Date'] = pd.to_datetime(coffee_data['Date'])
coffee_data['Year_Month'] = coffee_data['Date'].dt.to_period('M')
monthly_revenue = coffee_data.groupby('Year_Month')['money'].mean()
monthly_revenue.index = monthly_revenue.index.to_timestamp()
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Monthly Revenue")
```

**Analysis: Monthly Revenue Trend**
The monthly revenue trend reveals noticeable seasonal fluctuations. Revenue remains relatively stable from March to June, followed by a sharp decline during July and August. A recovery is observed from September onwards, with stabilization in the later months. The pattern suggests possible seasonality effects rather than consistent linear growth over time.

# PREDICTING NEXT MONTHS REVENUE


```python
monthly_df = monthly_revenue.reset_index()
monthly_df.columns = ['Month', 'Revenue']
monthly_df = monthly_df.sort_values('Month').reset_index(drop=True)

monthly_df['Month_Index'] = range(len(monthly_df))

print(monthly_df)
```

```python
n = len(monthly_df)

train = monthly_df[:9]
val = monthly_df[9:11]
test = monthly_df[11:]


print("Train shape:", train.shape)
print("Validation shape:", val.shape)
print("Test shape:", test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


#defining x and y
x_train = train[['Month_Index']]
y_train = train['Revenue']
model = LinearRegression()
model.fit(x_train,y_train)

#validation
val_pred = model.predict(val[["Month_Index"]])
#validation metrics
mae_val = mean_absolute_error(val['Revenue'],val_pred)
rmse_val = np.sqrt(mean_squared_error(val['Revenue'],val_pred))
r2_val = r2_score(val['Revenue'],val_pred)

print("validation mae : ",mae_val)
print("validation RMSE : ",rmse_val)
print("validaation R2 : ",r2_val)
```

**Validation Results Explanation**

The model was evaluated on the validation set using MAE, RMSE, and R score.
The MAE and RMSE values are around 1.9, which means the models predictions differ from the actual monthly revenue by about 1.9 units on average. Since the revenue values range roughly between 28 and 34, this error is relatively small and suggests that the model is able to capture the overall trend reasonably well.

However, the R score is highly negative. This is mainly because the validation set contains only two months of data. With such a small number of observations, the R metric becomes unstable and unreliable. Therefore, the R value in this case does not accurately reflect the true performance of the model.
